## <font color = 'yellow' >คำสั่งกราฟ 

```python
import matplotlib.pyplot as plt
#ข้อมูลแกน x , แกน y โดยต้องเป็นข้อมูลแบบชุด เช่น List , Numpy 
##Array , DataSeries และ DataFrame
x = [1,2,3,5] # ตำเเหน่งของข้อมูล เเกน X
y = [1,2.5,3.5,3] # ตำเเหน่งของข้อมูล เเกน y
plt.bar(x,y) #กราฟแท่ง
plt.show() #แสดงกราฟ
```

## <font color='yellow'>กราฟเส้น

```python
product1=[1.2,2.8,1.3,4.3,3.5]##ยอดขายสินค้า 1
product2=[2,1.4,3.3,3.1,1.2]; ## ยอดขายสินค้า 2
month=[1,2,3,4,5]; ##เดือนที่
plt.plot(month,product1) ## เส้นสีน้ำเงิน
plt.plot(month,product2) ## เส้นสีส้ม
plt.show()
```

###  <font color='pink'> การใส่ชื่อให้เเกน x เเละ y

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```
